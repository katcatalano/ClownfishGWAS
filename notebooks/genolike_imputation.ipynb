__calculate likelihoods with appropriate filters__

```python
#in the DNA directory-

!angsd -bam /local/home/katrinac/ClownfishGWAS/data/APPC_sequencing/DNA/trimmed_reads/mapping/genotypes/bamlist.list -GL 1 -nThreads 40 -doMajorMinor 1 -doMaf 1 -doCounts 1 -doPost 1 -minQ 20 -minMapQ 20 -SNP_pval 1e-6 -minInd 16 -minMAF 0.01  -doHWE 1 -minHWEpval 0.05 -setMaxDepth 115 -doGlf 2 -out ~/ClownfishGWAS/data/APPC_sequencing/DNA/trimmed_reads/mapping/genotypes/angsd/appc_angsd   
```
