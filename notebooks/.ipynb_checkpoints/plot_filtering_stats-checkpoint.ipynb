```r
library(tidyverse)
```

## Site level stats

```r
#read in vcftools output with data
#site quality
var_qual <- read_delim("~/ClownfishGWAS/data/APPC_sequencing/DNA/filtering/max_missing05_minQ20.lqual", delim = "\t",
           col_names = c("chr", "pos", "qual"), skip = 1)
summary(var_qual$qual)
a <- ggplot(var_qual%>% filter(qual <100), aes(qual)) + geom_density(fill = "dodgerblue1", colour = "black", alpha = 0.3)
a + theme_light()
```

```r
#mean depth for each variant
var_depth <- read_delim("~/ClownfishGWAS/data/APPC_sequencing/DNA/filtering/max_missing05_minQ20.ldepth.mean", delim = "\t",
           col_names = c("chr", "pos", "mean_depth", "var_depth"), skip = 1)
summary(var_depth$mean_depth)
a <- ggplot(var_depth, aes(mean_depth)) + geom_density(fill = "dodgerblue1", colour = "black", alpha = 0.3)
a + theme_light()

b <- ggplot(var_depth %>% filter(mean_depth <50), aes(mean_depth)) + geom_density(fill = "dodgerblue1", colour = "black", alpha = 0.3)
b + theme_light()
```

```r
# proportion of missingness at each variant
var_miss <- read_delim("~/ClownfishGWAS/data/APPC_sequencing/DNA/filtering/max_missing05_minQ20.lmiss", delim = "\t",
                       col_names = c("chr", "pos", "nchr", "nfiltered", "nmiss", "fmiss"), skip = 1)
summary(var_miss$fmiss)
a <- ggplot(var_miss, aes(fmiss)) + geom_density(fill = "dodgerblue1", colour = "black", alpha = 0.3)
a + theme_light()
```

```r
#minor allele frequency
var_freq <- read_delim("~/ClownfishGWAS/data/APPC_sequencing/DNA/filtering/max_missing05_minQ20.frq", delim = "\t",
                       col_names = c("chr", "pos", "nalleles", "nchr", "a1", "a2"), skip = 1)

# find minor allele frequency
var_freq$maf <- var_freq %>% select(a1, a2) %>% apply(1, function(z) min(z))

#plot 
a <- ggplot(var_freq, aes(maf)) + geom_density(fill = "dodgerblue1", colour = "black", alpha = 0.3)
a + theme_light()
```

## Individual level stats

```r
#mean individual depth
ind_depth <- read_delim("~/ClownfishGWAS/data/APPC_sequencing/DNA/filtering/max_missing05_minQ20.idepth", delim = "\t",
                        col_names = c("ind", "nsites", "depth"), skip = 1)
summary(ind_depth$depth)
a <- ggplot(ind_depth, aes(depth)) + geom_histogram(fill = "dodgerblue1", colour = "black", alpha = 0.3)
a + theme_light()
```

```r
#individual missingness
ind_miss  <- read_delim("~/ClownfishGWAS/data/APPC_sequencing/DNA/filtering/max_missing05_minQ20.imiss", delim = "\t",
                        col_names = c("ind", "ndata", "nfiltered", "nmiss", "fmiss"), skip = 1)

a <- ggplot(ind_miss, aes(fmiss)) + geom_histogram(fill = "dodgerblue1", colour = "black", alpha = 0.3)
a + theme_light()
```

```r
#individual heterozygosity
ind_het <- read_delim("~/ClownfishGWAS/data/APPC_sequencing/DNA/filtering/max_missing05_minQ20.het", delim = "\t",
           col_names = c("ind","ho", "he", "nsites", "f"), skip = 1)

a <- ggplot(ind_het, aes(f)) + geom_histogram(fill = "dodgerblue1", colour = "black", alpha = 0.3)
a + theme_light()
```
