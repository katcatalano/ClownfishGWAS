__calculate likelihoods and probs with appropriate filters__

```python
#in the genotype directory-

!angsd -bam /data/katrinac/APPC_sequencing/DNA/trimmed_reads/mapping/genotypes/bamlist.list -GL 1  -doMajorMinor 1 -doMaf 1 -doCounts 1 -doPost 1 -minQ 20 -minMapQ 20 -SNP_pval 1e-6 -minInd 16 -minMAF 0.01 -setMaxDepth 115  -nThreads 40 -beagleProb 1 -doGlf 2 -out /data/katrinac/APPC_sequencing/DNA/trimmed_reads/mapping/ensembl/genotypes/angsd/ensembl_appc_angsd_probs    
```

### Run ngsld with this file, then come back here to restrict angsd to only those sites. Can run pcangsd using the site filter for unlinked sites, without waiting for angsd re-run

__To restrict to unlinked sites with angsd, need to create a sorted sites file, and index with angsd__

```python
!angsd sites index /data/katrinac/APPC_sequencing/DNA/trimmed_reads/mapping/ensembl/genotypes/angsd/ngsld/appc_unlinked_for_angsd.txt
```

__angsd has trouble with the sites filter since not all sites are in it, so need to restrict to regions as well__

```python
#!zcat /data/katrinac/APPC_sequencing/DNA/trimmed_reads/mapping/genotypes/angsd/appc_angsd_probs.beagle.gz | head
!cut -f1 /data/katrinac/APPC_sequencing/DNA/trimmed_reads/mapping/ensembl/genotypes/angsd/ngsld/appc_unlinked_for_angsd.txt | sort|uniq >/data/katrinac/APPC_sequencing/DNA/trimmed_reads/mapping/ensembl/genotypes/angsd/ngsld/appc_unlinked_for_angsd.rf
#angsd -sites yourfile -rf your.rf
```
